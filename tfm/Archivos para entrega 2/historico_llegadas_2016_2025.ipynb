```python
import pandas as pd

# 1. Lista con los nombres de tus archivos CSV en orden INVERSO (De más nuevo a más viejo)
archivos = [
    '2025-2024.csv',
    '2023-2022.csv',
    '2021-2020.csv',
    '2019-2018.csv',
    '2017-2016.csv'
]

lista_dataframes = []
for archivo in archivos:
    try:
        df = pd.read_csv(archivo)

        # Nos aseguramos de que el bloque interno también esté de más nuevo a más viejo
        df = df.sort_values(by='Mes', ascending=False)

        lista_dataframes.append(df)
        print(f"Archivo cargado y ordenado inversamente: {archivo}")
    except FileNotFoundError:
        print(f"Error: No se encontró el archivo '{archivo}'.")

# 2. Concatenar todos los DataFrames verticalmente
if len(lista_dataframes) == len(archivos):
    df_historico_completo = pd.concat(lista_dataframes, ignore_index=True)

    # Por seguridad, aplicamos un ordenamiento global descendente por la columna 'Mes'
    df_historico_completo = df_historico_completo.sort_values(by='Mes', ascending=False).reset_index(drop=True)

    # 3. Guardar el resultado final
    nombre_salida = 'historico_llegadas_2025_2016.csv'
    df_historico_completo.to_csv(nombre_salida, index=False, encoding='utf-8-sig')

    # Mostrar las primeras y últimas filas para verificar el orden invertido
    print("Primeras filas (Debería empezar en 2025-12):")
    display(df_historico_completo.head(3))

    print("\nÚltimas filas (Debería terminar en 2016-01):")
    display(df_historico_completo.tail(3))
```

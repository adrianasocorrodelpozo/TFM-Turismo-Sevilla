```python
import pandas as pd

# Nombre exacto de tu archivo del INE
archivo_viajeros = 'ine_viajeros_ocupacion_clean.csv'

try:
    # 1. Cargar el histórico de llegadas del aeropuerto (2025-2016)
    df_llegadas = pd.read_csv('historico_llegadas_2025_2016.csv')
    df_llegadas.columns = df_llegadas.columns.str.strip()
    df_llegadas['Mes'] = df_llegadas['Mes'].astype(str).str.strip()

    # 2. Cargar tu archivo de viajeros y ocupación del INE
    df_viajeros = pd.read_csv(archivo_viajeros)
    df_viajeros.columns = df_viajeros.columns.str.strip()

    # 3. Buscar la columna 'Periodo' (sin importar mayúsculas/minúsculas)
    columna_periodo = None
    for col in df_viajeros.columns:
        if 'periodo' in col.lower():
            columna_periodo = col
            break

    if columna_periodo is None:
        raise KeyError(f"No se encontró la columna 'Periodo'. Columnas disponibles: {df_viajeros.columns.tolist()}")

    # Renombramos temporalmente a 'Mes' para que coincida con las llegadas al hacer el merge
    df_viajeros = df_viajeros.rename(columns={columna_periodo: 'Mes'})

    # Aseguramos formato de texto limpio 'YYYY-MM'
    df_viajeros['Mes'] = df_viajeros['Mes'].astype(str).str.extract(r'(\d{4}-\d{2})')[0]

    # 4. Merge
    df_consolidado = pd.merge(df_viajeros, df_llegadas, on='Mes', how='outer')

    # 5. Ordenamos de forma descendente estricta (De más nuevo 2025-12 a más viejo 2016-01)
    df_consolidado = df_consolidado.sort_values(by='Mes', ascending=False).reset_index(drop=True)

    # Devolvemos el nombre 'Periodo' a la columna de fecha
    df_consolidado = df_consolidado.rename(columns={'Mes': 'Periodo'})

    # 6. Guardar el archivo maestro
    nombre_salida = 'Dataset_Nadine_2016_2025.csv'
    df_consolidado.to_csv(nombre_salida, index=False, encoding='utf-8-sig')

    # Mostrar las columnas separadas en pantalla
    pd.set_option('display.max_columns', None)
    pd.set_option('display.expand_frame_repr', False)
    pd.set_option('display.width', 1000)

    # Vista previa del resultado en columnas reales independientes
    display(df_consolidado.head(5))
```
